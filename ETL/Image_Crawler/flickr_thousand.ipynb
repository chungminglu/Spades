```python
from pymongo import MongoClient
import gridfs
db = MongoClient().gridfs_project_image
thousand = gridfs.GridFSBucket(db)

import shutil # 把資料copy到檔案裡
import concurrent.futures

with concurrent.futures.ThreadPoolExecutor(max_workers=5) as executor: # 高併發
    for i in range(1,101):    
        url = "https://api.flickr.com/services/rest?sort=relevance&parse_tags=1&content_type=7&extras=can_comment%2Ccount_comments%2Ccount_faves%2Cdescription%2Cisfavorite%2Clicense%2Cmedia%2Cneeds_interstitial%2Cowner_name%2Cpath_alias%2Crealname%2Crotation%2Curl_c%2Curl_l%2Curl_m%2Curl_n%2Curl_q%2Curl_s%2Curl_sq%2Curl_t%2Curl_z&per_page=100&page={}&lang=zh-Hant-HK&text={}&viewerNSID=&method=flickr.photos.search&csrf=&api_key={}&format=json&hermes=1&hermesClient=1&reqId=c40b0214&nojsoncallback=1".format(i,"千本鳥居","6f7ab37efbd7a27f514e98f80173b36d")
        res = requests.get(url)
        res.text
        img_lish = res.json()['photos']['photo']
        imgurl=[]

        for image in img_lish:
            try:
                image['url_m_cdn']
                imgurl.append(image['url_m_cdn'])
            except:
                pass
        for img in imgurl:
            try:
                fname = img.split("/")[-1]
                print(fname)
                imgsave = requests.get(img, stream = True) 
                a = thousand.put(imgsave.raw, filename = fname) # 創建檔案，並且將image寫入
                shutil.copyfileobj(imgsave.raw, a) # 將imgsave裡的資料by raw copy到a之中
                del imgsave # 把暫存資料清空
            except:
                pass
```
