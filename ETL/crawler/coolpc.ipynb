## 爬取手機板coolpc

```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
res = requests.post("http://coolpc.com.tw/m/m-list.php",data = {"G":7})
```

```python
res
```

```python
import lxml
soup = BeautifulSoup(res.text,'lxml')
```

```python
ssds = soup.select("span.Q > table")[0].select("td")
```

```python
ssds
```

```python
data = [row.text for row in ssds]
```

```python
result = []
for row in data :
    ssd = {}
    ssd["brand"] = row.split(" ")[0]
    warranty = re.findall('【(\w年)', row)
    ssd['warranty'] = warranty[0] if warranty else ""
    flash = re.findall('\wLC', row)
    ssd['flash'] = flash[0] if flash else ""
    read = re.findall('讀:(\d+)M', row)
    ssd['read'] = int(read[0]) if read else 0
    write = re.findall('寫:(\d+)M', row)
    ssd['write'] = int(write[0]) if write else 0
    cap = re.findall('(\d+)(G|TB)', row)[0]
    
    if cap[1] == 'TB':
        size = int(cap[0]) * 1000
    else:
        size = int(cap[0])
    ssd['cap'] = size 
    price = re.findall('\$(\d+)', row)
    ssd['price'] = int(price[0]) if price else 0
    result.append(ssd)
result
```

```python
import csv
```

```python
with open("ssd.csv", 'w') as f:
    w = csv.DictWriter(f, result[0].keys())
    w.writeheader()
    w.writerows(result)
```
