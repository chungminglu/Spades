```python
import shutil # copy
for i in range(1,101):
    url = "https://api.flickr.com/services/rest?sort=relevance&parse_tags=1&content_type=7&extras=can_comment%2Ccount_comments%2Ccount_faves%2Cdescription%2Cisfavorite%2Clicense%2Cmedia%2Cneeds_interstitial%2Cowner_name%2Cpath_alias%2Crealname%2Crotation%2Curl_c%2Curl_l%2Curl_m%2Curl_n%2Curl_q%2Curl_s%2Curl_sq%2Curl_t%2Curl_z&per_page=100&page={}&lang=zh-Hant-HK&text={}&viewerNSID=&method=flickr.photos.search&csrf=&api_key=562311a86df8cbd6341b4306e46ce39b&format=json&hermes=1&hermesClient=1&reqId=c40b0214&nojsoncallback=1".format(i,"")
    res = requests.get(url)
    res.text
    img_lish = res.json()['photos']['photo']
    imgurl=[]
    
    for image in img_lish:
        try:
            image['url_m_cdn']
            imgurl.append(image['url_m_cdn'])
        except:
            pass
    for img in imgurl:
        try:
            fname = img.split("/")[-1]
            print(fname)
            imgsave = requests.get(img, stream = True) 
            f = open("D:\Project_Image\\{}".format(fname), 'wb') # w binary
            shutil.copyfileobj(imgsave.raw, f) # imgsaveby raw copyf
            f.close # 
            del imgsave # 
        except:
            pass
```
